```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
usedcars = pd.read_csv('outliers_cleaned_usedcars.csv')

# Display the first few rows to understand the structure of the data
print(usedcars.head())

# Select numeric columns only
numeric_columns = usedcars.select_dtypes(include=['float64', 'int64']).columns

# Calculate descriptive statistics for numeric columns
numeric_stats = usedcars[numeric_columns].describe()
print("Descriptive Statistics for Numerical Features:")
print(numeric_stats)

# Calculate the mode for all columns
modes = usedcars.mode().iloc[0]
print("\nMode of each column:")
print(modes)

# Calculate skewness and kurtosis for numeric columns only
skewness = usedcars[numeric_columns].skew()
kurtosis = usedcars[numeric_columns].kurt()

print("\nSkewness of Numeric Columns:")
print(skewness)

print("\nKurtosis of Numeric Columns:")
print(kurtosis)

# Data Visualization

# Scatter plot: Price vs. Kms Driven
plt.figure(figsize=(10, 6))
sns.scatterplot(data=usedcars, x='Kms Driven', y='price', hue='Fuel Type')
plt.title('Price vs. Kms Driven')
plt.show()

# Scatter plot: Price vs. Year of Manufacture
plt.figure(figsize=(10, 6))
sns.scatterplot(data=usedcars, x='Year of Manufacture', y='price', hue='Fuel Type')
plt.title('Price vs. Year of Manufacture')
plt.show()

# Histogram for the 'price' column
plt.figure(figsize=(10, 6))
sns.histplot(usedcars['price'], bins=30, kde=True)
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Histograms for other numerical features
usedcars[numeric_columns].hist(bins=30, figsize=(20, 15))
plt.suptitle('Distribution of Numerical Features')
plt.show()

# Box plot: Price by Fuel Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='Fuel Type', y='price', data=usedcars)
plt.title('Price Distribution by Fuel Type')
plt.show()

# Box plot: Price by Transmission
plt.figure(figsize=(10, 6))
sns.boxplot(x='Transmission', y='price', data=usedcars)
plt.title('Price Distribution by Transmission')
plt.show()

# Correlation Heatmap
correlation_matrix = usedcars[numeric_columns].corr()

plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin=-1, vmax=1)
plt.title('Correlation Matrix with Focus on Price')
plt.show()

# Pairplot to visualize relationships between price and other features
sns.pairplot(usedcars, y_vars='price', x_vars=['Kms Driven', 'Year of Manufacture', 'Engine', 'Mileage'])
plt.suptitle('Pairplot of Price with Other Features', y=1.02)
plt.show()

# Bar plot: Average price by Fuel Type
plt.figure(figsize=(10, 6))
sns.barplot(x='Fuel Type', y='price', data=usedcars, estimator='mean')
plt.title('Average Price by Fuel Type')
plt.show()

# Bar plot: Average price by Transmission
plt.figure(figsize=(10, 6))
sns.barplot(x='Transmission', y='price', data=usedcars, estimator='mean')
plt.title('Average Price by Transmission')
plt.show()
```

General Observations:
Majority of Cars and Popular OEMs:

The most common body type (bt) is Hatchback, and the most frequent Original Equipment Manufacturer (OEM) is Maruti, indicating a preference for compact and economical cars in this dataset.

Most cars have Third Party insurance, which is the minimum required by law. This could reflect a cost-conscious approach among car owners.
Numerical Features:
Central Tendency:

The average price of cars in the dataset is approximately ₹6.87 lakhs, with a wide range from ₹1 lakh to ₹25.75 lakhs, suggesting diverse price points from budget to premium cars.
The average mileage is around 19.53 kmpl, which is typical for many cars in the Indian market.
Car Age and Usage:

The cars in the dataset are relatively new, with an average year of manufacture around 2016-2017.
The average kilometers driven (Kms Driven) is around 58,802 km, indicating moderate usage, with some cars being driven as much as 222,996 km.
Distribution and Skewness:
Skewness:

OwnerNo and Price are positively skewed, indicating that there are more cars with lower owner counts and lower prices, but a few cars with high values skew the average.
Features like Seats show significant positive skewness, which is expected as most cars have around 4 to 5 seats, with very few having more than 5 seats.
The extreme skewness in Miscellaneous_Miscellaneous_Gear Box suggests there might be a few outlier vehicles with a much higher gear box number, possibly indicating incorrect data or special vehicle types.
Kurtosis:

High kurtosis in features like Seats and Miscellaneous_Miscellaneous_Gear Box indicates a peaked distribution with more outliers, which could be due to a majority of cars having standard configurations, while a few cars deviate significantly.
The extremely high kurtosis in Miscellaneous_Miscellaneous_Gear Box suggests a non-normal distribution with a long tail, possibly influenced by outliers.
Mode and Common Values:
Frequent Characteristics:
The most common number of cylinders is 4, which is standard for many cars.
The most frequent engine type is an In-Line Engine, typical for most vehicles in this category.
Vehicle Dimensions and Weight:
Dimensions:
The typical car length in the dataset is around 4 meters, width around 1.7 meters, and height around 1.56 meters, aligning with the dimensions of compact to mid-size cars.
The kerb weight is centered around 1137 kg, which is standard for the types of vehicles represented.
Considerations for Further Analysis:
Outliers:

Given the high skewness and kurtosis in certain columns, especially Miscellaneous_Miscellaneous_Gear Box, and Seats, careful consideration should be given to outlier detection and treatment, as these may impact model performance.
The distribution of Year of Manufacture is slightly negatively skewed, indicating more recent cars are included, with fewer older vehicles. This should be considered when analyzing the car's age and its impact on price or other factors.
Potential Errors or Anomalies:

The data points with very high gear boxes or outliers in other features might need to be reviewed for possible data entry errors or special cases like rare or unique vehicle models.
