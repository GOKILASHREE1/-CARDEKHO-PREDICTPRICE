```python
import pandas as pd
import json

def correct_json_format(json_str):
    try:
        json_str = json_str.replace("'", '"')  # Replace single quotes with double quotes
        json_str = json_str.replace('None', 'null')  # Replace Python None with JSON null
        json_data = json.loads(json_str)  # Load corrected JSON string
        return json_data
    except json.JSONDecodeError as e:
        print(f"Error decoding JSON: {e}")
        print(f"Problematic data: {json_str}")
        return {}

def extract_top_keys_values(json_obj):
    if not json_obj or 'top' not in json_obj:
        return {}
    top_entries = json_obj['top']
    extracted = {}
    for entry in top_entries:
        key = entry.get('key')
        value = entry.get('value')
        if key:
            extracted[key] = value
        elif value:
            extracted[value] = value
    return extracted

# Load your DataFrame
df = pd.read_excel(r'C:\CARPRICEPREDICTION\virtualenvironment\bangalore_cars.xlsx')

# Correct JSON formatting for each relevant column
for col in ['new_car_detail', 'new_car_overview', 'new_car_feature', 'new_car_specs']:
    df[col] = df[col].apply(correct_json_format)

# Normalize each JSON column
df_detail = df['new_car_detail'].apply(pd.Series)
df_overview = df['new_car_overview'].apply(extract_top_keys_values).apply(pd.Series)
df_feature = df['new_car_feature'].apply(extract_top_keys_values).apply(pd.Series)
df_specs = df['new_car_specs'].apply(extract_top_keys_values).apply(pd.Series)

# Combine all normalized DataFrames into a single DataFrame
df_combined = pd.concat([df, df_detail, df_overview, df_feature, df_specs], axis=1)

# Add new column 'city' with value 'bangalore' and move it to the first position
df_combined.insert(0, 'city', 'bangalore')

# Drop original JSON columns if needed
df_combined = df_combined.drop(columns=['new_car_detail', 'new_car_overview', 'new_car_feature', 'new_car_specs'])

# Save to CSV
df_combined.to_csv('newbangalore_output_file.csv', index=False)
print("File saved successfully")
```

```python
import pandas as pd
import os

# Define the directory containing the CSV files
directory = r'C:\CARPRICEPREDICTION\virtualenvironment'  

# List all CSV files in the directory
csv_files = [f for f in os.listdir(directory) if f.endswith('.csv')]

# Initialize an empty list to hold DataFrames
df_list = []

# Loop through the CSV files and read them into DataFrames
for file in csv_files:
    file_path = os.path.join(directory, file)
    df = pd.read_csv(file_path)
    df_list.append(df)

# Concatenate all DataFrames into one
combined_df = pd.concat(df_list, ignore_index=True)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv('newcombined_dataset.csv', index=False)

print("CSV files concatenated successfully!")
```
